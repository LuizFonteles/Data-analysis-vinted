Transform every size string into a waist size for pants

```python
def transform(array):
    array["size"]=array["size"].replace("W","",regex=True)
    array["size"]=array["size"].replace("4XL","42",regex=True)
    array["size"]=array["size"].replace("XXXL","40",regex=True)
    array["size"]=array["size"].replace("XXL","38",regex=True)
    array["size"]=array["size"].replace("XL","36",regex=True)
    array["size"]=array["size"].replace("XS","30",regex=True)
    array["size"]=array["size"].replace("L","34",regex=True)
    array["size"]=array["size"].replace("M","33",regex=True)
    array["size"]=array["size"].replace("S","32",regex=True)
```
